```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_csv_data = pd.read_csv('Index2018.csv')
raw_csv_data
```

```python
raw_csv_data.date = pd.to_datetime(raw_csv_data.date, dayfirst = True)
```

```python
raw_csv_data
```

```python
raw_csv_data.set_index('date', inplace = True)
```

```python
raw_csv_data
```

```python
df_comp = raw_csv_data.copy()
```

```python
df_comp = df_comp.drop(['dax', 'ftse', 'nikkei'], axis = 1)
df_comp
```

```python
df_comp['market_value'] = df_comp.spx
```

```python
del df_comp['spx']
```

```python
df_comp
```

```python
df_comp.market_value.plot(figsize = (20, 5), title = "S&P 500")
plt.legend()
plt.show()
```

```python
rw = pd.read_csv('RandWalk.csv')
rw
```

```python
rw.date = pd.to_datetime(rw.date, dayfirst = True)
rw
```

```python
rw.set_index('date', inplace = True)
```

```python
rw
```

```python
rw.price.plot(figsize = (20,5), title = "RandomWalk")
df_comp.market_value.plot(figsize = (20,5), title = 'S&P500')
plt.legend()
plt.show()
```
